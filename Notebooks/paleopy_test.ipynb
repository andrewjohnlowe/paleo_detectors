```python
from paleo.paleopy_classes import *
from matplotlib import pyplot as plt

import matplotlib as mpl
mpl.rcParams.update({'font.size': 14,'font.family':'serif'})
```

### Zabuyelite

```python
Zab = Mineral("Zabuyelite")
Zab.showProperties()
Zab.showSRIM()
```

```python
x_bins = np.logspace(0,3,200)

plt.figure()

plt.title("Zabuyelite")

plt.loglog(x_bins[:-1], Zab.dRdx(x_bins, 1e-45, 5), label=r'$5\,\mathrm{GeV}$')
plt.loglog(x_bins[:-1], Zab.dRdx(x_bins, 1e-45, 50), label=r'$50\,\mathrm{GeV}$')
plt.loglog(x_bins[:-1], Zab.dRdx(x_bins, 1e-45, 500), label=r'$500\,\mathrm{GeV}$')
#plt.loglog(x_bins[:-1], Zab.dRdx(x_bins, 1e-45, 5000), label=r'$5000\,\mathrm{GeV}$')

plt.loglog(x_bins[:-1], Zab.dRdx_nu(x_bins), linestyle='--',label=r'Neutrinos')

plt.loglog(x_bins[:-1], Zab.fission_bkg(x_bins, T=1e9), linestyle=':', label='Fission')

plt.legend(fontsize=12)

plt.ylabel("dR/dx [1/nm/kg/Myr]")
plt.xlabel("x [nm]")
plt.ylim(1e-4,1e10)
plt.xlim(1,1000)

plt.savefig("../plots/Zabuyelite_spectra.pdf",bbox_inches="tight")

plt.show()
```
